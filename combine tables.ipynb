```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
tmp = pd.ExcelFile('FXdata.xlsx')
sheet_names = tmp.sheet_names
sheets = {name: pd.read_excel('FXdata.xlsx', sheet_name=name, skiprows=2) for name in sheet_names}
print(sheets)
```

```python
table_dict_dates = defaultdict(list)
table_dict_spot_bid = defaultdict(list)
table_dict_spot_ask = defaultdict(list)
table_dict_spot = defaultdict(list)
table_dict_1m_bid = defaultdict(list)
table_dict_1m_ask = defaultdict(list)


for ridx in range(len(sheets[sheet_names[0]])):
    contain_nan = False
    for name in sheet_names:
        if np.isnan(sheets[name]['Spot Bid'][ridx]):
            #print('in row %d, %s is nan'%(ridx, name))
            contain_nan = True
            break
    if not contain_nan:
        # if we get until here all currencies have a spot bid price at that date
        # write new row
        for idx, name in enumerate(sheet_names):
            curr = sheets[name]
            if idx == 0:
                table_dict_dates['Date'] += [curr['Start'][ridx]]

            table_dict_spot_bid[name + ' ' + 'Spot Bid'] += [curr['Spot Bid'][ridx]]
            table_dict_spot_ask[name + ' ' + 'Spot Ask'] += [curr['Spot Ask'][ridx]]
            table_dict_spot[name + ' ' + 'Spot'] += [curr['Spot'][ridx]]
            table_dict_1m_bid[name + ' ' + '1M Forward Bid'] += [curr['1M Forward Bid'][ridx]]
            table_dict_1m_ask[name + ' ' + '1M Forward Ask'] += [curr['1M Forward Ask'][ridx]]

df_dates = pd.DataFrame.from_dict(table_dict_dates)
df_spot_bid = pd.DataFrame.from_dict(table_dict_spot_bid)
df_spot_ask = pd.DataFrame.from_dict(table_dict_spot_ask)
df_spot = pd.DataFrame.from_dict(table_dict_spot)
df_1m_bid = pd.DataFrame.from_dict(table_dict_1m_bid)
df_1m_ask = pd.DataFrame.from_dict(table_dict_1m_ask)
```

```python
df_spot_bid = pd.concat([df_dates, df_spot_bid], axis=1, join='inner') 
df_spot_ask = pd.concat([df_dates, df_spot_ask], axis=1, join='inner') 
df_spot = pd.concat([df_dates, df_spot], axis=1, join='inner') 
df_1m_bid = pd.concat([df_dates, df_1m_bid], axis=1, join='inner') 
df_1m_ask = pd.concat([df_dates, df_1m_ask], axis=1, join='inner') 
```

```python
df_spot_bid.to_excel('spot_bid.xlsx')
df_spot_ask.to_excel('spot_ask.xlsx') 
df_spot.to_excel('spot.xlsx') 
df_1m_bid.to_excel('1m_bid.xlsx') 
df_1m_ask.to_excel('1m_ask.xlsx') 
```
